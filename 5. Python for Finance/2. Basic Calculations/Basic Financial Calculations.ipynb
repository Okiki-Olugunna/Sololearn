```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
```

```python
# how much do you have to invest today to get a future value of 1,000 8 years from today at a rate of 10%
npf.pv(rate = 0.10, nper=8, pmt=0, fv=1000)
```

```python
# calculate how much we have to pay monthly to pay back a loan of 100,000 in 5 years
# The yearly interest rate is 7%, and is calculated monthly

# As we need to calculate monthly payments, we set the rate and periods (nper) in months 
# The pv parameter shows the present value of the loan, while the fv shows the future value we want to achieve 

npf.pmt(rate=0.07/12, nper=5*12, pv=100000, fv=0) 
```

```python
# the monthly deposits needed to achieve 50000 in 5 years with a 10% annual interest rate: 

npf.pmt(rate=0.10/12, nper=5*12, pv=0, fv=50000) 
```

```python
# the monthly investment needed to have a future value of 1M in 2.5 yaesr at a 200% interest rate

npf.pmt(rate=2/12, nper=2.5*12, pv=0, fv=1000000)
```

```python
np.round(npf.nper(rate=0.5/12, pmt=2000*12, pv=0, fv=1000000), 5)
```

```python
# Let's assume we invested 5000 and got the following payments back: 500, 700, 1000, 3000.
# To calculate the IRR, we first need to declare an array with the values, with the first value being our initial investment:

cashflow = [-5000, 500, 700, 1000, 3000]

# Now, we can use the irr() function:
npf.irr(cashflow)

# this answer is annually, not overall
```

```python
# comparing the IRR of 2 different investments: 

cf1 = [-50000, 10000, 25000, 25000, 35000, 42000] 
cf2 = [-30000, 10000, 13000, 18000, 25000, 20000] 

print("Option 1: ", npf.irr(cf1)) 
print("Option 2: ", npf.irr(cf2)) 
```

```python
# plotting the revenue of a company

rev = [18000, 25000, 20000, 45000, 19500] 

plt.plot(rev)
plt.title("Revenue")
plt.savefig("revenue of random company")
```
