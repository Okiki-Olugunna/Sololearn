```python
# my code:

class Enemy:
  name = ""
  lives = 0
  def __init__(self, name, lives):
    self.name = name
    self.lives = lives

  def hit(self):
    self.lives -= 1
    if self.lives <= 0:
       print(self.name + ' killed')
    else:
        print(self.name + ' has '+ str(self.lives) + ' lives')

class Monster(Enemy): #added inheritance 
  def __init__(self):
    super().__init__('Monster', 3)

class Alien(Enemy): #added inheritance
  def __init__(self):
    super().__init__('Alien', 5)
  


m = Monster()
a = Alien()

while True:
    x = input()
    if x == "gun": #gun attacks monster
        m.hit()
    elif x == "laser": #laser attacks alien
        a.hit()
    elif x == 'exit':
        break
```
